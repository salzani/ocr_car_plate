```python
import pytesseract
from pytesseract import Output
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('itens/images/trecho_livro.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

# **Reading with pytesseract**

```python
config_tesseract = '--tessdata-dir "/usr/share/tesseract-ocr/4.00/tessdata" --psm 6'  
txt = pytesseract.image_to_string(img, lang = 'por', config=config_tesseract)
print(txt)
```
