```python
import pytesseract
from pytesseract import Output
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

# **Extracting text from license plates**

```python
img = cv2.imread('itens/images/placa_carro1.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)

# * IMG color manipulation -> BGR to GRAY
```

```python
config_tesseract = '--tessdata-dir "/usr/share/tesseract-ocr/4.00/tessdata"'

txt = pytesseract.image_to_string(gray_img, lang='por', config=config_tesseract)

print(txt) 

# ! Tesseract OCR did not work in this case, we need to treat the image better.
```
